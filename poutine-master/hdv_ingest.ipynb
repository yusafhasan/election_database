### querying our tables ###

1. We first look for the 10 most popular candidates in Harris County, TX for the 2018 House elections,
as determined by popular vote. This gives us an insight into how effectively candidates in the Houston area were able to 
excite their supporters to vote.

```python
%%bigquery 
select * from hdv_staging.House2018 WHERE state = 'TX' and jurisdiction = 'Harris'
ORDER BY votes DESC
limit 10
```

2. Next, we compare how 2018 Senate candidates performed in Travis County, TX. This gives us insight into how the Austin area voted.

```python
%%bigquery 
select * from hdv_staging.Senate2018 WHERE state = 'TX' and jurisdiction = 'Travis'
ORDER BY votes DESC
```

3. For the 2016 House elections, we see which 10 counties Democrats pulled the most votes in. This shows which areas in Maine are Democrat strongholds.

```python
%%bigquery 
select * from hdv_staging.House2016 WHERE state = 'ME' and party = 'Democrat'
ORDER BY votes DESC
limit 10
```

4. Here, we see which counties independent candidates performed best in, by percentage, in the 2016 Senate elections. Interestingly, the top 10 performances are all in Arizona!

```python
%%bigquery 
select * from hdv_staging.Senate2016 WHERE party = 'Independent'
ORDER BY votes/total_votes DESC
limit 10
```

5. Lastly, we see which Florida counties Donald Trump earned the most votes. Florida is always a swing state, so it is interesting to see which regions fared well for the incumbent.

```python
%%bigquery 
select * from hdv_staging.President2016 WHERE candidate = 'Donald Trump' and state = 'FL'
ORDER BY votes DESC
limit 10
```
